## Problema de Regresi√≥n

```python
import matplotlib.pyplot as plt
%matplotlib inline  
import numpy as np
from sklearn import datasets, linear_model

X = np.array(range(100))
y = 2*X+5 + np.random.normal(0,10,100)
plt.scatter(X,y)
```

```python
regresor = linear_model.LinearRegression()
lm = regresor.fit(X.reshape(-1,1), y.reshape(-1,1))
plt.scatter(X,y)
plt.plot(X, lm.predict(X.reshape(-1,1)), color='red', linewidth=4)
```
